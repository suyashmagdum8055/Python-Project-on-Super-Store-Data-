```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sb 
```

```python
dataset = pd.read_excel('Superstore_USA.xlsx')
```

```python
dataset.head(5)
```

```python
dataset.isnull().sum()
```

```python
dataset.shape
```

```python
dataset['Product Base Margin'] = dataset['Product Base Margin'].fillna(dataset['Product Base Margin'].mean())
```

# Product Category

```python
plt.figure(figsize=(9,6))
sb.countplot(x = 'Product Category',data=dataset[dataset["Product Category"]== "Office Supplies"],hue = "Product Sub-Category")
plt.show()
```

```python
plt.figure(figsize=(9,6))
sb.countplot(x = 'Product Category',data=dataset[dataset["Product Category"]== "Technology"],hue = "Product Sub-Category")
plt.show()
```

```python
plt.figure(figsize=(9,6))
sb.countplot(x = 'Product Category',data=dataset[dataset["Product Category"]== "Furniture"],hue = "Product Sub-Category")
plt.show()
```

```python
dataset['Order Year'] = dataset['Order Date'].dt.year
```

# "Customer Segment"

```python
plt.figure(figsize=(6,4))
sb.countplot(x = 'Customer Segment',data=dataset,color="gold")
plt.show()
```
